If real policy documents are not yet downloaded, the headwind module falls back to a synthetic time series so the notebook runs end to end.

````python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.nlp.headwind import compute_headwind_index

headwind_df = compute_headwind_index(use_synthetic=True)
print(headwind_df.tail(6).to_string(index=False))
````

````python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(headwind_df['month'], headwind_df['headwind_index'], marker='o', color='#2c3e50')
ax.axhline(0.6, color='red', linestyle='--', linewidth=0.8, label='High headwind threshold')
ax.fill_between(headwind_df['month'], headwind_df['headwind_index'], alpha=0.1, color='#2c3e50')
ax.set_ylim(0, 1)
ax.set_xlabel('Month')
ax.set_ylabel('Policy headwind index')
ax.set_title('Monthly policy headwind index (synthetic)')
ax.legend()
plt.xticks(rotation=35)
plt.tight_layout()
plt.savefig('../logs/headwind_index.png', bbox_inches='tight')
plt.show()
````

````python
# Demo NER on a sample policy sentence without requiring downloaded documents
import spacy

try:
    nlp = spacy.load('en_core_web_sm')
    sample = (
        'The Ministry of Labour and Employment has issued guidelines under the Code on Social Security '
        'to extend e-Shram registration benefits to gig workers in Dharavi, Maharashtra, with a '
        'monthly stipend of Rs 3,000 for reskilling under the NSDC scheme.'
    )
    doc = nlp(sample)
    print('Entities found in sample policy sentence:')
    for ent in doc.ents:
        print(f'  {ent.text!r:40s}  [{ent.label_}]')
except OSError:
    print('spaCy model not installed. Run: python -m spacy download en_core_web_sm')
````
